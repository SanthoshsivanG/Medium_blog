```python
import pandas as pd   

path = "./User_Data.csv"
X_full = pd.read_csv(path)

# Check dataset
X_full.head()
```

```python
X_full.shape
```

```python
X = X_full.copy()
```

```python
y = X.Purchased
y.shape
```

```python
X.drop(['User ID'], axis=1, inplace = True)
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train_full, X_valid_full, y_train, y_valid = train_test_split ( X,y,train_size = 0.8, test_size=0.2, random_state =0)
```

```python
X_train_full.dtypes
```

```python
#As there are columns with object datatype, that is not int or float, we want to convert it to numerical so that logistic regression can be done.
#select numerical columns

numerical_cols = [cname for cname in X_train_full.columns if X_train_full[cname].dtype in ['int64','float64']]
print(numerical_cols)
categorical_cols = [cname for cname in X_train_full.columns if X_train_full[cname].nunique() < 5 and X_train_full[cname].dtype =='object']
print(categorical_cols)
```

```python
my_col = numerical_cols + categorical_cols
X_train = X_train_full[my_col].copy()
X_valid = X_valid_full[my_col].copy()
```

```python
#one hot encoding using pandas

X_train_encoded = pd.get_dummies (X_train)
X_valid_encoded = pd.get_dummies (X_valid)

#X_train_encoded.shape = (320,5)
#X_valid_encoded.shape =(80,5)
```

```python
# To prevent maximum iteration issues of model
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()

X_train_encoded = sc_X.fit_transform(X_train_encoded)
X_valid_encoded = sc_X.transform(X_valid_encoded)
```

```python
# Fitting the Logistic Regression into the Training set
from sklearn.linear_model import LogisticRegression

logr = LogisticRegression(random_state = 0)
logr.fit(X_train_encoded, y_train)
```

```python
prediction = logr.predict(X_valid_encoded)
prediction
```

```python
from sklearn.metrics import log_loss

log_error = log_loss(y_valid,prediction)
print(f"Log Loss (Log Error): {log_error:.4f}")
```

```python
from sklearn.metrics import accuracy_score

accuracy =accuracy_score(y_valid, prediction)
accuracy
```
